```python
Total_Players_df=players_df.groupby('Country')[['Player_Name']].count()
Total_Players_df
```

```python
plt.plot(Total_Players_df,'x--r');
plt.xlabel('country');
plt.ylabel('Total Players');
plt.xticks(rotation=45);
plt.title('country wise total players');
sns.set_style("darkgrid")
plt.rcParams["figure.figsize"] = (10,10)
```

```python
plt.bar(Total_Players_df.index.values,Total_Players_df['Player_Name'])
plt.xlabel('country');
plt.ylabel('Total Players');
plt.xticks(rotation=45);
plt.title('country wise total players');
```

```python
df=Total_Players_df.sort_values(by=['Player_Name'],ascending=False)
df
```

```python
plt.bar(df.index.values,df['Player_Name'])
plt.xlabel('country');
plt.ylabel('Total Players');
plt.xticks(rotation=45);
plt.title('country wise total players');
```

```python
Total_wins_df=matches_df.groupby('winner')[['winner']].count()
Total_wins_df
```

```python
plt.barh(Total_wins_df.index.values,Total_wins_df['winner'])
plt.ylabel('IPL Team');
plt.xlabel('Total Wins');
plt.title('Team wise total wins');
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(Total_wins_df['winner'],labels=Total_wins_df.index.values,startangle=90,autopct='%1.1f%%');
plt.title('Team wise total wins',color='blue');
ax1.axis('equal');
plt.tight_layout()
```

```python
teams_df=pd.read_csv('teams.csv')
print(teams_df)
```

```python
teams_wise_df=pd.read_csv('teamwise_home_and_away.csv')
teams_wise_df
```

```python
X_axis = np.arange(len(teams_wise_df.team))
plt.barh(X_axis+0.2,teams_wise_df.home_win_percentage,0.3,color='orange',label='home win %');
plt.barh(X_axis-0.2,teams_wise_df.away_win_percentage,0.3,color='green',label='away win %');
plt.xlabel('winning percentage')
plt.yticks(X_axis, teams_wise_df.team)
plt.ylabel('IPL teams')
plt.title('IPL team home and away winning %')
plt.legend();



```

```python
most_runs_df=pd.read_csv('most_runs_average_strikerate.csv')
most_runs_df
```

```python
mruns_df=most_runs_df.sort_values(by='total_runs').tail(10)
mruns_df
```

```python
plt.barh(mruns_df.batsman,mruns_df.total_runs);
plt.xlabel('Player Name')
plt.ylabel('Runs')
plt.title('Most Runs in IPL',color='brown');
```

```python
srate_df=most_runs_df[most_runs_df.numberofballs>500].sort_values(by='strikerate',ascending=False).head(20)
srate_df
```

```python
sns.barplot(y='batsman',x='strikerate',data=srate_df);
plt.title("Players with best strikerate ",color='blue')
```

```python
team=input('please enter name of the team: ')
season_wise_w_trend_df=matches_df[matches_df.winner==team].groupby(['Season'])[['winner']].count();
team_df=matches_df[(matches_df.team1==team)| (matches_df.team2==team)]
season_wise_l_trend_df=team_df[matches_df.winner!=team].groupby(['Season'])[['winner']].count()
plt.plot(season_wise_l_trend_df,'x--r', label='loses');
plt.plot(season_wise_w_trend_df,'s--g', label='wins');
plt.xticks(rotation=45);
plt.ylabel('No of wins/loses')
plt.title('IPL year wise trend of {}'.format(team),color='blue');
sns.set_style("darkgrid")
plt.legend();
```

```python
fig,axes=plt.subplots(2,2)
plt.tight_layout(pad=3.5)
plt.rcParams["figure.figsize"] = (15,10)

axes[0,0].barh(mruns_df.batsman,mruns_df.total_runs);
axes[0,0].set(xlabel='Player Name', ylabel='Runs')
axes[0, 0].set_title('Most Runs in IPL',color='brown');


df=Total_Players_df.sort_values(by=['Player_Name'],ascending=False)
axes[0,1].bar(df.index.values,df['Player_Name'])
axes[0,1].set(xlabel='Country', ylabel='Total Players')
axes[0,1].set_xticklabels(labels=df.index.values, rotation=45)
axes[0,1].set_title('Country Wise Total Players',color='brown');


axes[1, 0].pie(Total_wins_df['winner'],labels=Total_wins_df.index.values,startangle=90,autopct='%1.1f%%');


X_axis = np.arange(len(teams_wise_df.team))
axes[1,1].barh(X_axis+0.2,teams_wise_df.home_win_percentage,0.3,color='orange',label='home win %');
axes[1,1].barh(X_axis-0.2,teams_wise_df.away_win_percentage,0.3,color='green',label='away win %');
axes[1,1].set(xlabel='winning percentage', ylabel='IPL teams')
axes[1,1].set_title('IPL team home and away winning %',color='brown');
axes[1,1].legend();

```
